```python
import pandas as pd
import numpy as py
import collections
from collections import Counter
from itertools import groupby
from operator import itemgetter
```

```python
HP=pd.read_csv('Harry Potter Data.csv')
HP=HP.dropna(subset=['Location'])
HPT=HP.T.to_dict().values()
for item in HPT:
    item['times']=len(item['Page'].split(','))

towhom=[]
```

```python
HP.head()
```

```python
def request(name):
    for item in HPT:
       
        if item['Character - Name']==name:
            tuple1 = (item['To Whom'], item['times'])
            print tuple1
            towhom.append(tuple1)
    return towhom
```

```python
request('Hermoine Granger')        

```

```python
d=collections.defaultdict(int)
a=[]
b=[]
for na,ti in towhom:
    if not na in a:a.append(na)
    d[na]+=ti
for f in a:
    tuple2=(f,d[f])
    print tuple2
    b.append(tuple2)
network=[]    
def newtuple(query):
    for i in range(len(b)):
        tuple3=(query,b[i][0],b[i][1])
        network.append(tuple3)
        
    return network
```

```python
newtuple('Hermoine Granger')
```

```python
import networkx as nx
G = nx.DiGraph()
G.add_weighted_edges_from(network)
import matplotlib.pyplot as plt
elarge=[(u,v) for (u,v,d) in G.edges(data=True) if d['weight'] >10]
esmall=[(u,v) for (u,v,d) in G.edges(data=True) if d['weight'] <=10]
pos=nx.spring_layout(G)
nx.draw_networkx_nodes(G,pos,node_size=700)
nx.draw_networkx_edges(G,pos,edgelist=elarge,
                    width=6)
nx.draw_networkx_edges(G,pos,edgelist=esmall,
                    width=6,alpha=0.5,edge_color='b',style='dashed')
nx.draw_networkx_labels(G,pos,font_size=10,font_family='sans-serif')
plt.axis('off')
plt.savefig("weighted_graph.png") # save as png
plt.show() 
```
